```python
cd_top10= cd_final.groupby('topic')['datavalue'].mean().to_frame().reset_index()
cd_top10.sort_values(by = 'datavalue',ascending = False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='topic', y='datavalue',data=cd_top10)
plt.title('Top 10 Chronic Disease')
plt.show();
```

```python
plt.figure(figsize=(10, 6))
plt.barh(cd_top10['topic'], cd_top10['datavalue'])
plt.xlabel('Data Value')
plt.ylabel('Chronic Disease')
plt.title('Top 10 Chronic Diseases by Average Data Value')
plt.gca().invert_yaxis()  # Invert y-axis to display highest value at the top
plt.show()
```

```python
cd_top10_by_year= cd_final.groupby(['yearstart','topic'])['datavalue'].mean().to_frame().reset_index()
cd_top10_by_year.sort_values(by=['yearstart', 'datavalue'], ascending=[True, False])
```

```python
plt.figure(figsize=(12, 8))
grouped = cd_top10_by_year.groupby(['yearstart', 'topic']).mean().unstack()
grouped.plot(kind='bar', stacked=False)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Data Value')
plt.title('Data Value of Chronic Diseases Over Time')
plt.legend(title='Topic', bbox_to_anchor=(1, .5), loc='center left')


# Show plot
#plt.tight_layout()
plt.show()
```

```python
age_group_stats = cd_age1.groupby(['stratification1', 'topic'])['datavalue'].mean().reset_index()
age_group_stats
```

```python
# Plotting
plt.figure(figsize=(10, 6))

# Iterate over unique chronic diseases and plot their rates by age group
for disease in age_group_stats['topic'].unique():
    subset = age_group_stats[age_group_stats['topic'] == disease]
    plt.plot(subset['stratification1'], subset['datavalue'], marker='o', label=disease)

plt.xlabel('Age Group')
plt.ylabel('Average Rate')
plt.title('Rates of Chronic Diseases by Age Group')
plt.legend(bbox_to_anchor=(1, .5), loc='center left')
#plt.grid(True)
plt.show()
```

```python
gender_stats = cd_final.groupby(['stratification1', 'topic'])['datavalue'].mean().reset_index()
gender_stats.sort_values(by = 'datavalue', ascending = False)
```

## corelation bw chronic disease and obesity with age >65 

```python
n_final_dropna = n_final.dropna(subset=['datavalue'])
ob_test_dropna = ob_test.dropna(subset=['data_value'])
```

```python
min_len = min(len(n_final_dropna), len(ob_test_dropna))
n_final_dropna = n_final_dropna[:min_len]
ob_test_dropna = ob_test_dropna[:min_len]
```

```python
corr_without_null = n_final_dropna['datavalue'].corr(ob_test_dropna['data_value'], method='pearson', min_periods=1)
corr_without_null
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(n_final_dropna['datavalue'], ob_test_dropna['data_value'], alpha=0.5)
plt.title('Corelation bw Chronic Disease and Obesity (age over 65)')
plt.xlabel('Chronic Disease datavalue ')
plt.ylabel('Obesity datavalue')
#plt.grid(True)

```
